```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Dataset

```python
df = pd.read_csv('Boston.csv')
df.head(10)
```

```python
df.drop(columns=['Unnamed: 15','Unnamed: 16'], inplace=True)
```

```python
df.drop(columns=['CAT. MEDV'], inplace=True)
```

#### Checking for null values

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

#### Checking correlation with target variable MEDV

```python
corr = df.corr()

plt.figure(figsize=(14,14))
sns.heatmap(corr, fmt=".2%", annot=True)
```

```python
df.corr()['MEDV'].sort_values()
```

```python
X = df.loc[:,['LSTAT','PTRATIO','RM']]
Y = df.loc[:,"MEDV"]
X.shape,Y.shape
```

### Preparing training and testing data set

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=10)
```

### Normalizing training and testing dataset

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

### Preparing model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(128,input_shape=(3,),activation='relu',name='input'))
model.add(Dense(64,activation='relu',name='layer_1'))
model.add(Dense(1,activation='linear',name='output'))
model.compile(optimizer='adam', loss='rmse', metrics=['accuracy'])

model.summary()
```

```python
hist = model.fit(x_train, y_train, epochs=100, validation_split=0.05)
```

```python
output = model.evaluate(x_test,y_test)
```

```python
print(f"Mean Squared Error: {output[0]}"
      ,f"Mean Absolute Error: {output[1]}",sep="\n")
```

```python
y_pred = model.predict(x=x_test)
```

```python
print("Predicted Y    Actual Y")
for y_p, y_t in zip(y_pred, y_test):
    print(y_p[0], "\t", y_t)
```

```python
plt.scatter(y_train, y_pred)
```
